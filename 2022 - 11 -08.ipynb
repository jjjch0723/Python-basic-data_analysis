```python
import csv
import re
import matplotlib.pyplot as plt

f = open("Busan_R.csv","r", encoding='utf-8')
data = csv.reader(f)
JinGu = []
HaeUnDaeGu = []
DongRaeGu = []
Yongdo = []

for row in data : 
    if '부산진구' in row[0]:
        for i in row[3:]:
            j = i.replace(',' , '')
            JinGu.append(int(j))

    
    if '해운대구' in row[0]:
        for i in row[3:]:
            j = i.replace(',', '')
            HaeUnDaeGu.append(int(j))
            
    if '동래구' in row[0]:
        for i in row[3:]:
            j = i.replace(',', '')
            DongRaeGu.append(int(j))
            
    if '영도' in row[0]:
        for i in row[3:]:
            j = i.replace(',', '')
            Yongdo.append(int(j))
            
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title("부산의 인구 분포도")

plt.plot(JinGu, 'ro-', label='부산진구')
plt.plot(HaeUnDaeGu, 'bx-', label='해운대구')
plt.plot(DongRaeGu, 'g^-', label='동래구')
plt.plot(Yongdo, 'ys-', label='영도')

plt.legend()
plt.show()
```
